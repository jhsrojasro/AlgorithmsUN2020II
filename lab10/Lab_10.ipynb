```
import random 
from matplotlib import pyplot as plt, animation 
from itertools import permutations
from math import factorial 

num_pasos_total = 0
num_pasos_por_perm = 0
N = 0

def swap(V, i, j): 
    V[j], V[j+1] = V[j+1], V[j]
  
def bubblesort(V):
    global num_pasos_total
    global num_pasos_por_perm
    n = len(V) 
    for i in range(n): 
        for j in range(0, n-i-1):
            num_pasos_total = num_pasos_total + 1
            num_pasos_por_perm = num_pasos_por_perm + 1
            if V[j] > V[j+1]: 
                swap(V, i, j)
                num_pasos_total = num_pasos_total + 1
                num_pasos_por_perm = num_pasos_por_perm + 1
  
def paint():
    global N 
    global num_pasos_por_perm
    N = 3
    perm = permutations(list(range(1, N + 1)))
    for i in list(perm):   
      V = list(i)
      fig, ax = plt.subplots() 
      ax.set_title("Bubble Sort") 
      bar_sub = ax.bar(range(len(V)), V, align="edge") 

      ax.set_xlim(0, N) 
      text = ax.text(0.02, 0.95, "", transform=ax.transAxes) 
      iteration = [0] 

      plt.show() 
      plt.close() 
      bubblesort(V) 
      print("Number of steps = ", num_pasos_por_perm)
      num_pasos_por_perm = 0

paint()
print("n = ", N)
print("Number of permutations = ", factorial(N)) 
print("Average number of steps =  ", num_pasos_total/(factorial(N)))
```

### Sucesión de Fibonnaci elaborada con una función recursiva:

```
from time import time

def fib_recursive(n):
  if (n==0 or n==1):
    return n
  else:
    return fib_recursive(n-1) + fib_recursive(n-2)
  
def calc_time(n):
  start = time()
  num = fib_recursive(n)
  interval = time() - start
  return interval
```

```
import matplotlib.pyplot as plt
import numpy as np
from timeit import timeit

times = np.arange(50)
for i in range(len(times)):
  times[i] = calc_time(i)
x = np.arange(0,50)
y = x
plt.plot(x,y)
plt.plot(x,times)
plt.show()
```

### Sucesión de Fibonnaci elaborada con una función iterativa:

```
from time import time

def fib_iterative(n):
  results = []
  siguiente = 1
  actual = 0
  temporal = 0
  for i in range(1,n):
    temporal = actual
    actual = siguiente
    siguiente= siguiente + temporal
  return results

def calc_time_iterative(n):
  start = time()
  num = fib_iterative(n)
  interval = time() - start
  return interval
```

```
import matplotlib.pyplot as plt
import numpy as np
from timeit import timeit

times = np.arange(50)
for i in range(len(times)):
  times[i] = calc_time_iterative(i)
x = np.arange(0,50)
y = x
plt.plot(x,y)
plt.plot(x,times)
plt.xlabel('n')
plt.ylabel('Tiempo de ejecución')
plt.title('Tiempo de ejecución sucesión de Fibonacci iterativo')
plt.show()
```
